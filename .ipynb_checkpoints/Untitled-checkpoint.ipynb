```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
df = pd.read_csv('fcc-forum-pageviews.csv', index_col = 0, parse_dates = True)
```

```python
#check if theres any null values
df.isna().values.any()
```

```python
df = df[ (df['value'] >= df['value'].quantile(q = 0.025)) & (df['value'] <= df['value'].quantile(q = 0.975)) ]
```

```python
df.shape
```

```python
fig, axs = plt.subplots(figsize=(12,6))
axs.plot(df.index, df['value'], color='orange', linewidth=1.5, label='orangeLine')
axs.set_xlabel("Date")
axs.set_ylabel("Page Views")
axs.legend()
axs.set_title("Daily freeCodeCamp Forum Page Views 5/2016-12/2019")
```

```python
df_pr = df
df_pr
```

```python
df["month"] = df.index.month
df["year"] = df.index.year
```

```python
df
```

```python
df_bar = df.groupby(['month','year'])['value'].mean()
```

```python
# try setting the indexes to ['month'] and 'year' and then unstack ?
```

```python
df_bar
```

```python
df_bar = df_bar.unstack(level=0)
```

```python
df_bar
```

```python
"""
I debugged it a little bit.

The DataFrame.pivot() and DataFrame.pivot_table() are different.
pivot() doesn't accept a list for index.
pivot_table() accepts.
Internally, both of them are using reset_index()/stack()/unstack() to do the job.

pivot() is just a short cut for simple usage, I think.
"""
```

```python
df_bar.plot(kind='bar')
```
